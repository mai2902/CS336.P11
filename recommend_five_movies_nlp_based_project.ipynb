- Here we have genre infor along with other unnecessary things so let's remove this first

```python
def convert(obj):
    L = []  # empty list
    for i in ast.literal_eval(obj): # ast.literal_eval() is used to convert string to list
        i['name']   # i is a dictionary
        L.append(i['name']) # append the name of the genre
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']
```

```python
# same thing applied on keyword as well
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
# Lets explore the crew column
movies['crew'][0]
```

```python
def director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head()
```

```python
movies['overview'] = movies.overview.apply(lambda x:x.split())
```

```python
# Remove the spaces between the words
movies['genres'] = movies.genres.apply(lambda x : [i.replace(" " , "") for i in x])
movies['keywords'] = movies.keywords.apply(lambda x : [i.replace(" " , "") for i in x])
movies['cast'] = movies.cast.apply(lambda x : [i.replace(" " , "") for i in x])
movies['crew'] = movies.crew.apply(lambda x : [i.replace(" " , "") for i in x])
```

In NLP (Natural Language Processing) tasks, spaces between words are typically removed in the following situations:

**Tokenization:** During tokenization, words are split into individual tokens, and spaces are used to separate these tokens. However, in some cases, you might perform further processing where spaces are removed to normalize or standardize the text.

**Text Normalization:** In preprocessing steps like normalization, spaces might be removed to standardize text. For example, multiple consecutive spaces might be reduced to a single space, or trailing and leading spaces might be trimmed.

**Text Cleaning:** When cleaning text data, unnecessary spaces are often removed to ensure consistency and improve the quality of the data. This includes removing extra spaces between words, as well as leading and trailing spaces.

**Concatenation:** In some advanced tasks, you might concatenate words or phrases into a single string where spaces are removed to create identifiers or keys, especially in tasks like creating unique IDs or merging multiple features into a single string.

**String Matching or Search:** When performing string matching or search operations, removing spaces might be useful to improve matching accuracy or to normalize strings for comparison.

In general, spaces are managed according to the specific requirements of the task and the preprocessing pipeline. It's essential to handle spaces appropriately to maintain the integrity and meaning of the text data.

    <h2>TEXT VECTORIZATION (BAG OF WORDS)</h2>
</div>

Text Vectorization is the process of converting text data into numerical format so that it can be used in machine learning algorithms. One popular method for text vectorization is the Bag of Words (BoW) model.

**Bag of Words (BoW) Model**
The Bag of Words model is a simple and widely-used technique in text processing and natural language processing (NLP) for converting text into numerical vectors. Heres how it works:

**Tokenization:** Split the text into individual words or tokens. For example, the sentence "I love NLP" would be tokenized into ["I", "love", "NLP"].

**Vocabulary Creation:** Create a list of all unique words (the vocabulary) from the entire corpus of text. For instance, if the corpus includes "I love NLP" and "NLP is great", the vocabulary would be ["I", "love", "NLP", "is", "great"].

**Vectorization:** Represent each document as a vector based on the vocabulary. Each position in the vector corresponds to a word in the vocabulary, and the value is typically the count of the word in the document (or sometimes the term frequency-inverse document frequency, TF-IDF, score).

Example:
Consider two sentences:

"I love NLP"
"NLP is great"
Vocabulary:
["I", "love", "NLP", "is", "great"]

**Vectorization:**

For "I love NLP":

Count of "I": 1
Count of "love": 1
Count of "NLP": 1
Count of "is": 0
Count of "great": 0
Vector: [1, 1, 1, 0, 0]
For "NLP is great":

Count of "I": 0
Count of "love": 0
Count of "NLP": 1
Count of "is": 1
Count of "great": 1
Vector: [0, 0, 1, 1, 1]

**Key Points:**
Simplicity: BoW is easy to understand and implement.
Sparsity: The resulting vectors are often sparse because not all words appear in every document.
Loss of Context: BoW does not capture the order or context of words, which can be a limitation for some NLP tasks.

**Applications:**
Text Classification: Categorizing text into predefined categories.
Spam Detection: Identifying spam emails based on their content.
Sentiment Analysis: Determining the sentiment expressed in a piece of text.
Despite its simplicity, the Bag of Words model is a foundational technique in text analysis and can be very effective for many NLP tasks.

```python
# Ignore the Stopwords
cv = CountVectorizer(max_features=5000, stop_words='english') # max_features is used to take top 5000 words, stop_words is used to ignore the common words like 'is', 'the', 'a', 'an' etc
```

```python
# Fit the tags into the CountVectorizer
vectors = cv.fit_transform(df['tags']).toarray()
vectors[0]
```

    <h2>STEMMING</h2>
</div>

Stemming is a text preprocessing technique in natural language processing (NLP) that reduces words to their root or base form. The goal of stemming is to normalize words so that different forms of a word are treated as the same word, which helps in improving the performance of various NLP tasks.

**How Stemming Works**
Stemming algorithms remove prefixes or suffixes from words to achieve a common root form. For example:

"running"  "run"

"flies"  "fli" (Note: This might not always be a meaningful stem)

"fished"  "fish"


```python
ps = PorterStemmer()
```

```python
def Stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return  " ".join(y)
```

```python
new_df = df[df['tags'].notna()].copy()
new_df['tags'] = new_df['tags'].apply(Stem)
```

```python
%%writefile app.py
import streamlit as st
import pickle
import pandas as pd
import requests

def fetch_poster(movie_id):
    response = requests.get(f'https://api.themoviedb.org/3/movie/{movie_id}?api_key=475cd503f2e7c81659655de2a1e23ad9&language=en-US')
    data = response.json()
    return f"https://image.tmdb.org/t/p/w500/{data['poster_path']}"

def fetch_overview(movie_id):
    response = requests.get(f'https://api.themoviedb.org/3/movie/{movie_id}?api_key=475cd503f2e7c81659655de2a1e23ad9&language=en-US')
    data = response.json()

    if 'overview' in data:
        return data['overview']
    else:
        return "Overview not available."

def recommend(movie, num_recommendations):
    movie_index = movies[movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:num_recommendations+1]

    recommended_movies = []
    recommended_movies_posters = []
    recommended_movies_overviews = []
    for i in movies_list:
        movie_id = movies.iloc[i[0]].id

        recommended_movies.append(movies.iloc[i[0]].title)
        recommended_movies_posters.append(fetch_poster(movie_id))
        recommended_movies_overviews.append(fetch_overview(movie_id))
    return recommended_movies, recommended_movies_posters, recommended_movies_overviews

movies_dict = pickle.load(open('movies_dict.pkl', 'rb'))
movies = pd.DataFrame(movies_dict)

similarity = pickle.load(open('similarity.pkl', 'rb'))

st.title(" MOVIE RECOMMENDATION SYSTEM ")

st.markdown(
    """
    <style>
    .title {
        font-size: 24px;
        font-weight: bold;
        color: #ff6347;
    }
    .dropdown {
        font-size: 16px;
        color: #4682b4;
    }
    </style>
    """,
    unsafe_allow_html=True
)

st.markdown('<div class="title"> Discover Your Next Favorite Movie! </div>', unsafe_allow_html=True)

selected_movie_name = st.selectbox(
    'Ready to Discover Your Next Favorite Film? Choose a Movie and Let the Magic Begin! ',
    movies['title'].values,
    format_func=lambda x: f" {x}"
)

num_recommendations = st.slider(
    'How many recommendations would you like? ',
    min_value=1,
    max_value=10,
    value=5
)

if st.button('Get Recommendations! '):
    with st.spinner('Fetching recommendations...'):
        names, posters, overviews = recommend(selected_movie_name, num_recommendations)

        cols = st.columns(num_recommendations)
        for i, col in enumerate(cols):
            with col:
                st.image(posters[i], use_container_width=True)
                st.text(names[i])
                st.write(overviews[i])
```
